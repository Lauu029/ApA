```python
#Cálculo del límite de la función mediante operaciones entre vectores
import numpy as np
import matplotlib.pyplot as plt

#Función que voy a integrar
def funcion(x):
    return (x*x) + 5*x -3

#puntos límite de la funcion
lim1 = 0
lim2 = 10

#Genero la funcion con puntos x e y
fun_x = np.linspace(lim1, lim2, 100)
fun_y = funcion(fun_x)

#Escribo la función en la consola
print("f(x)= x^2 + 5x -3")

#Cálculo del punto máximo de la funcion
M= np.max(fun_y)
print("Máximo de la función: ",M)

#Para pintar la función y ver de forma gráfica los resultados
def pintaFuncion(fun_x,fun_y,puntos_x,puntos_y):
    # Crea el gráfico
    plt.plot(fun_x, fun_y)
    plt.grid(True)  # Muestra una cuadrícula en el gráfico
    plt.scatter(puntos_x, puntos_y, color='red', s=3)
    # Muestra el gráfico
    plt.show()
    
# integral de la función(mediante operaciones con vectores)
def integra_mc(fun,a,b,num_puntos):
    #Genero los puntos aleatorios
    pos_x = np.random.uniform(a, b, num_puntos)
    pos_y = np.random.uniform(0, M, num_puntos)
    debajo = pos_y <= fun(pos_x)
    puntos_debajo_x = pos_x[debajo]
    puntos_debajo_y = pos_y[debajo]
    pintaFuncion(fun_x, fun_y, puntos_debajo_x, puntos_debajo_y)
    return np.mean(debajo) * (b - a) * M
#imprimo la solución de la integral
solucion= integra_mc(funcion, lim1,lim2,10000)
print("Resultado de la integral: ", solucion)
```
