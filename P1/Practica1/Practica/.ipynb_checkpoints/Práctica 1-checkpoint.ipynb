```python
from utils import load_data
from linear_reg import gradient_descent
from linear_reg import compute_gradient
from linear_reg import compute_cost

import matplotlib.pyplot as plt
import numpy as np

puntos_x, puntos_y  = load_data()
```

```python
#Ver visualmente los datos de dataset
plt.scatter(puntos_x, puntos_y, color="blue", s=3)
plt.title("Puntos del dataset")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```

```python
from public_tests import compute_cost_test
compute_cost_test(compute_cost)
```

```python
from public_tests import compute_gradient_test
compute_gradient_test(compute_gradient)
```

```python
x = puntos_x
y = puntos_y

w_initial = 0
b_initial = 0

alpha = 0.01
num_iters = 1500

w_optimal, b_optimal, J_history = gradient_descent(x, y, w_initial, b_initial, compute_cost, compute_gradient, alpha, num_iters)

from public_tests import compute_gradient_descent_test
compute_gradient_descent_test(w_optimal,b_optimal)
```

```python
plt.plot(np.arange(num_iters), J_history)
plt.xlabel('Iteraciones')
plt.ylabel('Coste')
plt.title('Gradient Descent')
plt.show()
```

```python
#Dibujar la recta de regresión
plt.scatter(x, y, color="blue", s=3)
plt.plot(puntos_x, w_optimal * puntos_x + b_optimal, color='red', label="Recta de Regresión")
plt.title("Recta de Regresión")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.show()
```
